```python
import pandas as pd
import numpy as np
```

```python
#import files
all_topics = 'alltopics.csv'
posts = 'postsforanalysis.csv'
users = 'usersforanalysis.csv'

all_topics_df = pd.read_csv(all_topics)
posts_df = pd.read_csv(posts, low_memory = False)
users_df = pd.read_csv(users)
```

```python
#removes NaN values in Posts_DF
posts_df = posts_df.fillna(False)
#posts_df.loc[posts_df['wordpress'], 'name']
```

```python
#confirms that the columns are now boolean columns
posts_df['wordpress'].unique()
```

```python
user_posts_df = posts_df.merge(users_df, on = 'user_id')
list(user_posts_df)
```

```python
#testing if an item belongs to multiple class
posts_df.iloc[0, 12:312]
#based on this, some item does belong to multiple classes
```

```python
#getting the name of the product and creating DF of the number of times 
#a person upvotes a certain item type
item_types = list(posts_df.iloc[:, 12:312].columns.values)

user_upvote_category = pd.DataFrame([])

for item_type in item_types:
    temp = user_posts_df.loc[user_posts_df[item_type], ['name_y', 'name_x']].groupby('name_y').count()
    temp.columns = [item_type]
    temp = temp.reset_index()
    #print(temp.head())
    
    if user_upvote_category.empty:
        user_upvote_category = user_upvote_category.append(temp)
    else:
        user_upvote_category = user_upvote_category.merge(temp, on = 'name_y', how = 'outer')
    
```

```python
#create mean max and std
user_upvote_category = user_upvote_category.fillna(0)
user_upvote_category['mean'] = user_upvote_category.mean(axis = 1)
user_upvote_category['max'] = user_upvote_category.max(axis = 1)
user_upvote_category['std'] = user_upvote_category.std(axis = 1)

user_upvote_category.sort_values(['max', 'mean'], ascending = False)

user_upvote_category_sorted = user_upvote_category.iloc[:, 0:301]
```

```python
user_upvote_category.to_csv('user and category preference.csv')
```
