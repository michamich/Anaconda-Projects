```python
import pandas as pd
import numpy as np
```

```python
result_path = 'data\college result.csv'
KIE_path = 'data\kauffman startup index.csv'
college_df = pd.read_csv(result_path, encoding = 'utf-8')
KIE_df = pd.read_csv(KIE_path)
#college_df.head()
```

```python
KIE_df['Category'].unique()
```

```python
state_bool = (KIE_df['Category'] == 'Large') | (KIE_df['Category'] == 'Small')
year_state_bool = state_bool & (KIE_df['Index Year'] == 2015)
#KIE_df.loc[state_bool, 'Location_name' ].unique()

KIE_filtered = KIE_df.loc[year_state_bool, :]
```

```python
KIE_filtered.sort_values('Location_name').head()
KIE_filtered['cur_index'] = KIE_filtered['index_last_y'] + KIE_filtered['index_diff']
#KIE_filtered.loc[:, ['cur_index', 'Location_name']].sort_values('cur_index', ascending = False)

KIE_sorted = KIE_filtered.loc[:, ['cur_index', 'Location_name']].sort_values('cur_index', ascending = False)
KIE_sorted.columns = ['cur_index', 'state']
```

```python
def above75perc (index):
    if index > 0.57:
        return True
    else:
        return False
    
KIE_sorted['above75'] = KIE_sorted['cur_index'].map(above75perc)
```
